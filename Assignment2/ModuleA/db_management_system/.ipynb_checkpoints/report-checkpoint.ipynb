## Introduction

Efficient data storage and retrieval are fundamental challenges in database systems. As datasets grow larger, performing operations using linear scans becomes inefficient. To address this problem, database systems use indexing structures that enable fast lookup operations.

One of the most widely used indexing structures is the **B+ Tree**. B+ Trees are balanced tree structures that support efficient insertion, deletion, search, and range queries. They are widely used in database management systems and file systems.

In this project, we implement a lightweight database indexing engine using a **B+ Tree** from scratch in Python. The implemented system supports storing records associated with keys and executing efficient database operations. The system is then applied to the ride-sharing dataset provided in the assignment.

Additionally, the performance of the B+ Tree is compared with a brute-force database approach to demonstrate the advantages of structured indexing.

## Dataset Description

The dataset provided for this assignment represents a ride-sharing database.

The dataset contains tables such as:

- Members
- ActiveRides

Each table is stored as a CSV file. Each row represents a record in the table.

The primary key column of each table is used as the index key in the B+ Tree.

During the execution of the notebook, all CSV files are automatically loaded and indexed using the B+ Tree structure.

## System Architecture

The implemented system consists of the following components:

### Database Manager
The DatabaseManager class manages all tables and their corresponding indexes.

### Table Layer
Each table stores records in memory using a dictionary structure.

### B+ Tree Index
The B+ Tree stores keys and references to records. Internal nodes guide the search process, while leaf nodes store the actual key-value pairs.

Leaf nodes are connected using linked pointers, which allows efficient execution of range queries.

```python
import os
import pandas as pd
import random
import time
import tracemalloc
import matplotlib.pyplot as plt

from IPython.display import Image, display

from database.db_manager import DatabaseManager
from database.bplustree import BPTree
from database.bruteforce import BruteForceDB
```

## Initialize Database Manager

The DatabaseManager is responsible for creating tables, loading records, and managing indexes.

```python
db = DatabaseManager()
```

## Load Dataset Tables

All CSV files present in the current directory are automatically detected and loaded.

For each table:

1. The table is created
2. The CSV file is loaded
3. A B+ Tree index is built on the primary key column

```python
for file in os.listdir():

    if file.endswith(".csv"):

        table_name = file.replace(".csv","")

        df = pd.read_csv(file)

        key_column = df.columns[0]

        table = db.create_table(table_name, key_column)

        table.load_csv(file)

        db.build_index(table_name)

        print("Loaded table:", table_name)
```

## Exact Search Operation

Exact search retrieves a specific record using its primary key.

The B+ Tree enables efficient lookup by navigating from the root node to the appropriate leaf node.

```python
db.search("Members", 28498441)
```

## Range Query

Range queries retrieve all records whose keys fall within a given interval.

Because leaf nodes in the B+ Tree are connected through linked pointers, range queries can be executed efficiently by scanning consecutive leaf nodes.

```python
db.range_query("Members", 28498450, 28498460)
```

## Insert Operation

Insertion adds a new record to the database and updates the B+ Tree index accordingly.

If the leaf node becomes full during insertion, the node is split and the tree structure is updated.

```python
new_member = {
    "MemberID":999999,
    "Name":"TestUser"
}

db.insert("Members",new_member)

db.search("Members",999999)
```

## Update Operation

```python
db.insert("Members",{"MemberID":999999,"Name":"TestUser"})
print("Before update:", db.search("Members",999999))

db.indexes["Members"].update(999999,{"MemberID":999999,"Name":"UpdatedUser"})
print("After update:", db.search("Members",999999))
```

## Delete Operation

Deletion removes a record from the database and updates the B+ Tree structure.

```python
db.delete("Members",999999)
db.search("Members",999999)
```

## Performance Analysis

The performance of the B+ Tree is compared with the BruteForceDB implementation.

We measure:

- insertion time
- search time
- deletion time
- range query time
- random workload performance
- memory usage

```python
sizes=list(range(1000,20000,3000))

insert_bpt=[]
insert_brute=[]

search_bpt=[]
search_brute=[]

delete_bpt=[]
delete_brute=[]

range_bpt=[]
range_brute=[]
```

```python
for size in sizes:

    keys=random.sample(range(1,size*10),size)

    search_keys=random.sample(keys,size//10)

    delete_keys=random.sample(keys,size//10)

    r_start=random.randint(1,size*5)
    r_end=r_start+(size//2)

    bpt=BPTree(order=5)
    brute=BruteForceDB()

    start=time.time()
    for k in keys:
        bpt.insert(k,k)
    insert_bpt.append(time.time()-start)

    start=time.time()
    for k in keys:
        brute.insert(k)
    insert_brute.append(time.time()-start)

    start=time.time()
    for k in search_keys:
        bpt.search(k)
    search_bpt.append(time.time()-start)

    start=time.time()
    for k in search_keys:
        brute.search(k)
    search_brute.append(time.time()-start)

    start=time.time()
    for k in delete_keys:
        bpt.delete(k)
    delete_bpt.append(time.time()-start)

    start=time.time()
    for k in delete_keys:
        brute.delete(k)
    delete_brute.append(time.time()-start)

    start=time.time()
    bpt.range_query(r_start,r_end)
    range_bpt.append(time.time()-start)

    start=time.time()
    brute.range_query(r_start,r_end)
    range_brute.append(time.time()-start)
```

```python
plt.figure(figsize=(12,8))

plt.subplot(221)
plt.plot(sizes,insert_bpt,label="B+ Tree")
plt.plot(sizes,insert_brute,label="Brute Force")
plt.title("Insertion Time")

plt.subplot(222)
plt.plot(sizes,search_bpt,label="B+ Tree")
plt.plot(sizes,search_brute,label="Brute Force")
plt.title("Search Time")

plt.subplot(223)
plt.plot(sizes,delete_bpt,label="B+ Tree")
plt.plot(sizes,delete_brute,label="Brute Force")
plt.title("Deletion Time")

plt.subplot(224)
plt.plot(sizes,range_bpt,label="B+ Tree")
plt.plot(sizes,range_brute,label="Brute Force")
plt.title("Range Query Time")

plt.legend()
plt.tight_layout()
plt.show()
```

## Random Workload Performance

```python
ops=["insert","search","delete"]

tree=BPTree(order=5)

keys=random.sample(range(100000),5000)

start=time.time()

for _ in range(2000):

    op=random.choice(ops)

    key=random.choice(keys)

    if op=="insert":
        tree.insert(key,key)

    elif op=="search":
        tree.search(key)

    else:
        tree.delete(key)

random_workload_time=time.time()-start

random_workload_time
```

## Memory Usage Analysis

```python
tracemalloc.start()

tree = BPTree(order=5)

for k in range(5000):
    tree.insert(k,k)

current, peak = tracemalloc.get_traced_memory()

print("Current memory usage:", current)
print("Peak memory usage:", peak)

tracemalloc.stop()
```

## Results

The benchmarking results show that the B+ Tree performs significantly better than the brute-force approach as dataset size increases.

Insertion, search, and deletion operations scale efficiently in the B+ Tree due to its logarithmic time complexity.

Range queries are particularly efficient in the B+ Tree because leaf nodes are linked sequentially.

The brute-force approach, which relies on linear scanning, becomes increasingly inefficient as dataset size grows.
